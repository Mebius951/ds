```python
remarks_base = [
    "Patient has chest-pain!",
    "No heart disease.",
    "High cholesterol, needs care.",
    "Normal ECG; stable.",
    "Risk-level: moderate!"
]
# Calculate how many times the base list needs to be repeated to cover len(df)
repeat_count = (len(df) // len(remarks_base)) + 1
# Create a list that is long enough
extended_remarks = remarks_base * repeat_count
# Slice the extended list to the exact length of the DataFrame
df['remarks'] = extended_remarks[:len(df)]

print("Text column BEFORE removing punctuation:")
print(df['remarks'].head())
```

```python
import string

df['remarks'] = df['remarks'].str.replace(
    f"[{string.punctuation}]", "", regex=True
)

print("\nText column AFTER removing punctuation:")
print(df['remarks'].head())
```
